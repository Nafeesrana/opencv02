# Draw Shapes in an Image

```python
#create an image using numpy
new_image = np.zeros(shape = (820, 1020, 3), dtype= np.int16)
```

```python
new_image.shape
```

```python
new_image
```

```python
plt.imshow(new_image)
```

```python
#draw rectangle on the image
cv2.rectangle(new_image, pt1 = (100, 90), pt2= (400, 300), color = (0,0,250), thickness=-1)
```

```python
plt.imshow(new_image)
```

```python
#Draw circle on this Image
cv2.circle(img=new_image, center= (300,400), radius=100, color=(250,0,0), thickness=-1)
```

```python
plt.imshow(new_image)
```

```python
cv2.line(new_image, pt1= (0,0), pt2= (1020,820), color = (102, 255, 250), thickness=10)
```

```python
plt.imshow(new_image)
```

# Create Text in an Image

```python
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
cv2.putText(new_image, text="Kesa Lga Mzaak", org=(20, 600), fontFace=font,
            fontScale=2, color=(250, 250, 250), thickness=3, lineType=cv2.LINE_AA)
```
